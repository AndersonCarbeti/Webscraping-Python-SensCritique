###Objectif: The project aims to collect data on movies of a user from the popular film website SensCritique.

###SensCritique is a French website dedicated to the culture of cinema, series, music, and books, it allows users to rate, review, and create collections of the media they've seen, read, and listen.

###The website will be accessed using the Python programming language, and the Selenium library will be utilized to automate the process of navigating the website and extracting the relevant information.
###The collected data will be futher analyzed to gain insights and make informed decisions.


```python
##1. Aims, objectives and background
##1.1. Introduction

#library
import csv
import time
import selenium
import bs4
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


    
# Set up the webdriver
driver = webdriver.Chrome()
driver.get('https://www.senscritique.com/Wanderson/collection')

# Click on the button
# Wait 20 seconds for the page to load
time.sleep(15)
try:
    button_xpath = '//button[@id="didomi-notice-agree-button"]'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()
except Exception as e:
    print("An error occurred:", e)
    
    # Click on the button
# Wait 20 seconds for the page to load
time.sleep(15)
try:
    button_xpath = '//*[@id="__next"]/div[1]/div/main/div[2]/div/div[1]/div[2]/div[2]/p[6]'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()
except Exception as e:
    print("An error occurred:", e)
    
    # Click on the button
# Wait 20 seconds for the page to load
time.sleep(15)
try:
    button_xpath = '//*[@id="__next"]/div[1]/div/main/div[2]/div/div[1]/div[3]/div[2]/p[2]'
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()
except Exception as e:
    print("An error occurred:", e)

no_of_pages = int(input('Enter the #pages to scrape: ')) - 6

# From page 1 to 5

for loop_counter in range(1, 6):
    xpath = '//*[@id="__next"]/div[1]/div/main/div[2]/div/div[2]/div/div[2]/nav/span[' + str(loop_counter) + ']'
    element = driver.find_element(By.XPATH, xpath)
    element.click()
    time.sleep(2)

    # Find the elements again after the page has been refreshed or the element has been clicked
    elements_a = driver.find_elements(By.TAG_NAME, 'a')
    elements_p = driver.find_elements(By.TAG_NAME, 'p')

    # Create a new file for each element
    with open('element_' + str(loop_counter) + '.txt', 'w') as f:
        # Write the element text to the file
        for element in elements_a + elements_p:
            f.write(element.text + '\n')



# Get the number of pages to scrape from the user
no_of_pages = int(input('Enter the #pages to scrape: ')) - 6

# Iterate the specified number of times
for i in range(no_of_pages):
    # Locate the "Next Page" button
    button = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div/main/div[2]/div/div[2]/div/div[2]/nav/span[6]')

    # Click on the button
    button.click()

    # Wait 2 seconds for the page to load
    time.sleep(2)

    # Find all the links and p elements on the page
    elements_a = driver.find_elements(By.TAG_NAME, 'a')
    elements_p = driver.find_elements(By.TAG_NAME, 'p')

    # Create a new file for each iteration
    with open('page_' + str(i+1) + '.txt', 'w') as f:
        # Write the element text to the file
        for element in elements_a + elements_p:
            f.write(element.text + '\n')

# Locate the "End" button
buttonend = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div/main/div[2]/div/div[2]/div/div[2]/nav/span[7]')

# Click on the button
buttonend.click()

# Find all the links on the page
elements = driver.find_elements(By.TAG_NAME, 'a')

# Create a new file for the last page
with open('page_' + str(no_of_pages+1) + '.txt', 'w') as f:
    # Write the element text to the file
    for element in elements:
        f.write(element.text + '\n')

```
